```python
from IPython.display import clear_output
import requests
import json
import time
# ==================================== Property Class ======================================================<Property>
class Property:
    def __init__(self,home_id, address, property_features, income, expence):
        self.address = address
        self.address["property"] = property_features
        self.address["income"] = income
        self.address["expence"] = expence
        self.property = {home_id:self.address}
    
    def __str__(self):
        return self.property
    
    
    
# ====================================== Adding new property =======================================< Property adress >
def property_input():
    valid = False
    while not valid:
        home_id = input("Enter Home Id ")
        
        # *************** check the property exist ******************
        check = check_id(home_id)
        if check != -1:
            print("Property exists")
            break
        #------------------------------------------------------------ 
        
        street = input("Enter street address ").upper()
        zip_code = input("zip code")
        property_features = {}
        address = {}
        income = {}
        expence = {}
        
        def validate_address():
            base_url ="https://api.lob.com/v1"
            api_method = "/us_verifications"
            Api_key = "live_82bd4dc5f202bc4752ee348d00f671baf71"
            url = base_url+api_method

            d= {'primary_line':street,'zip_code':zip_code}

            address_res = requests.post(url, data = d,auth=(Api_key,''))  
            return address_res.json()
    
        result = validate_address()
        if 'error' not in result and result['valid_address']:
            address["street"] = street
            address["city"] = result["components"]["city"]
            address["state"] = result["components"]["state"]
            address["zip_code"] = zip_code
            property_features["county"] = result["components"]["county"]
            property_features["address_type"]=result["components"]["address_type"]
            
            
            
# ===================================== property features ==========================================< property features >
            c = input("Press 1 -> to add property details or any button to skip")
            if c == '1':
                property_features["bedrooms"]    = input("Enter number of bedrooms ")
                property_features["bathrooms"]   = input("Enter number of bathrooms ")
                property_features["size"]        = input("Enter size Sq. Ft. ")
                property_features["year"]        = input("Enter year of built ")
                property_features["Description"] = input("Description ")
            
            
            
# ===================================== income info ==============================================< adding income info >
            income["rental"]    = input("Enter rental income $")
            income["laundry"]   = input("Enter laundry income $")
            income["storage"]   = input("Enter storage income $")
            while True:     # misc incomes 
                s = input("Press 1 -> to add more income or press any button to coninue ")
                if s=='1':
                    t = input("Enter type of income ")
                    a = input("Enter amount $")
                    income[t] = a
                else:
                    break
                

                
# ====================================== expence info ============================================< adding expenses >
            expence["Tax"]       = input("Enter tax $")
            expence["insurance"] = input("Enter incurace $")
            expence["HOA"]       = input("Enter HOA expence $")
            expence["lawn"]      = input("Enter lawn expence $")
            expence["vacancy"]   = input("Enter vacancy $")
            expence["repair"]    = input("Enter repair $")
            expence["copex"]     = input("Enter cop expences $")
            expence["management"]= input("Enter property management fee $")
            expence["mortgage"]  = input("Enter mortgage amout $")
            
            s = input("Press -> 1 to add utility expences or press any button to quit ")
            if s == '1':
                expence["utility"] = {}
                expence["utility"]["electricity"] = input("Enter electricity expence $")
                expence["utility"]["water"]       = input("Enter water expence $")
                expence["utility"]["sewer"]       = input("Enter sewer expence $")
                expence["utility"]["garbage"]     = input("Enter garbage expence $")
                expence["utility"]["gas"]         = input("Enter gas expence $")        
                while True:
                    s = input("Press 1 -> to add more expences or press any button to continue ")
                    if s=='1':
                        e = input("Enter type of expence ")
                        a = input("Enter amount $")
                        expence["utility"][e] = a
                    else:
                        break
            
            

# ============================== adding new recond into file =====================================< adding new property >
            new_property = Property(home_id, address, property_features, income, expence)
            add_new("property.json",new_property)
            print("The property has been added in out system")
            break
            
            
        else:
            print("Invalid address please try again ... ")
            q = input("press 1 to try again or press any button to quit ")
            if q != '1':
                break
                
                
                
# ====================================== calculate income expens Cash flow =============================================
def cash_flow():
    list_properties()
    home_id = input("Enter home id that you would want to invest ")
    check = check_id(home_id)
    if check!=-1:
        lst = read_file("property.json")
        home = lst[check]
        inc = income(home)
        exp = expence(home)
        print("income = ", inc)
        print("expence = ", exp)
        return inc-exp
    else:
        print("The proparty has not been found !")
        return -1
                
                
def income(lst):
    d = lst[list(lst.keys())[0]]["income"]
    total = 0
    for i in d:
        if d[i].isnumeric():
            total+=int(d[i])
    return total
                
def expence(lst):
    d = lst[list(lst.keys())[0]]["expence"]
    total = 0
    for i in d:
        if isinstance(d[i], str) and d[i].isnumeric():
            total+=int(d[i])
    
    if "utility" in d.keys():
        utilities = d["utility"]
        for u in utilities:
            if utilities[u].isnumeric():
                total+=int(utilities[u])
    return total
        
    
# ============================================== Cash on Cash ROI ======================================================
def roi():
    cf = cash_flow()
    if cf!= -1:
        r = {}
        r["down_payment"] = input("Enter down payment $")
        r["closing_cost"] = input("Enter closing cost $")
        r["rehab_budget"] = input("Enter Rehab budget $")
        while True:
            c = input("Press 1 -> to add more expence or press any button to quit")
            if c == "1":
                k = input("Enter type of expence ")
                v = input("Enter amount of cost$")
                r[k] = v
            else:
                break
        inv = invest(r)
        cocrio = ((cf*12)/inv)*100
        print(f"Cash on Cash ROI = ${cocrio:.2f}%")
        input("Press any button to continue ...")
    else:
        print("We encountered an issue try again ")
            
def invest(r):
    total = 0
    for i in r:
        if r[i].isnumeric():
            total+=int(r[i])
    return total
            
# ========================================== Helper functions ==========================================================                
# Check home id in the list ----------------------------------------------------------------------------
def check_id(home_id):
    lst = read_file("property.json")
    if len(lst):
        home_ids = list(key for x in lst for key in x)
        if home_id in home_ids:
            return home_ids.index(home_id)
    return -1                
                              
# listing properties --------------------------------------------------------------------
def list_properties():
    lst = read_file("property.json")
    if len(lst):
        print("================================ Propery list =================================")
        for p in lst:
            print(p)
            
# deleting propery ----------------------------------------------------------------------------------
def delete_property():
    list_properties()
    home_id = input("Enter property id that will be removed ")
    check = check_id(home_id)
    if check!=-1:
        lst = read_file("property.json")
        del lst[check]
        update("property.json",lst)
        print("The property has been removed ...")
        list_properties()
    else:
        print("The property not found")
                
# read files -----------------------------------------------------------------------------------------
def read_file(file):
        with open(file) as p:
            return json.load(p)
    
def add_new(file, new_record):
#     with open(file, 'a') as prop:
#         prop.write(new_record.writable())
    data = []    
    with open(file) as p:
        data = json.load(p)  
    data.append(new_record.__str__())
    
    with open(file, mode='w', encoding='utf-8') as p:
        json.dump(data, p)
        

def update(file, recent):
    with open(file, mode='w', encoding='utf-8') as p:
        json.dump(recent, p)
        
        
        
# -----------------------------------------------------------------------------------------------------
def menu():
    print("Press 1 -> add new property\n"+
          "Press 2 -> list properties\n"+
          "Press 3 -> remove property\n"+
          "Press 4 -> ROI calculate\n"+
          "Press 5 -> quit")
    
def main():
    while True:
        time.sleep(1)
        clear_output()
        menu()
        time.sleep(1)
        c = input("Enter your choice ")
        if c=="5":
            break
        elif c=="1":
            property_input()
        elif c=="2":
            list_properties()
            input("press any button to back to main menu")
        elif c=="3":
            delete_property()
        elif c=="4":
            roi()
main()
    
    
    
```
